```python
import matplotlib.pyplot as plt
import pandas as pd 
from datetime import datetime
```

```python
data = pd.read_csv('Turnos de aula_.csv')
```

```python
dados = {}
for indice, linha in data.iterrows():
    if linha[4] not in dados:
        curso = linha[4]
        dados[curso] = []

print(dados)
print(len(dados))
#linha: vetor com as info
#indice: qual linha
```

```python
## linha[4] : Curso de Graduação
## linha[6] : horario preferido ( Parte matutino do formulario )
## linha[10] : horario preferido ( Parte Diurno do formulario )
## linha[14] : horario preferido ( Parte Noturno do formulario )
quantidade_respostas = 0
print(len(data))

for indice, linha in data.iterrows():
    curso = linha[4]
    if linha[5] == 'Matutino' and str(linha[6]) != 'nan':
        dados[curso].append(linha[6])
    if linha[5] == 'Diurno' and str(linha[10]) != 'nan':
        dados[curso].append(linha[10])
    if linha[5] == 'Vespertino' and str(linha[10]) != 'nan':
        dados[curso].append(linha[10])
    if linha[5] == 'Noturno' and str(linha[14]) != 'nan':
        dados[curso].append(linha[14])

dados_unica = dados.copy()
print(dados_unica)
```

```python
print(dados['Engenharia de Sistemas'])
```

```python
# Contando a quantidade de respostas para garantir que nenhuma tenha sido deixada para trás:
quantidade_respostas = 0
quantidade_votantes_dict = {}
print("Quantidade de alunos que responderam o formulário: ")
for curso in dados_unica:
    ##len(dados_unica[curso]) -> quantidades de respostas
    quantidade_votantes_curso = len(dados_unica[curso])
    print(curso + ' ' + str(quantidade_votantes_curso) + " respostas")

    if curso in quantidade_votantes_dict:
        # Se existir, soma o valor à chave existente
        quantidade_votantes_dict[curso] += quantidade_votantes_curso
    else:
        # Se não existir, cria a chave com o valor desejado
        quantidade_votantes_dict[curso] = quantidade_votantes_curso

print("quantidade total de alunos que responderam: " + str(sum(quantidade_votantes_dict.values())))
```

```python
## Uma resposta pode ter mais que 1 horário preferido, essa parte do código garante que os dois sejam isolados em "2 respostas" diferentes
for curso in dados_unica:
    respostas = []
    for resposta in dados[curso]:
        resposta_temp = resposta.split(';')
        for horario in resposta_temp:
            if all(c.isdigit() or c == ':' or c == '-' or c == ' ' for c in horario) and len(horario) > 11:
                respostas.append(horario)
    dados[curso] = sorted(respostas)
```

```python
# Contando a quantidade de respostas para garantir que nenhuma tenha sido deixada para trás:
print("Quantidade de respostas: ")
quantidade_respostas_curso = 0
quantidade_respostas_dict = {}
for curso in dados:
    quantidade_respostas_curso = len(dados[curso])
    print(curso + " " + str(quantidade_respostas_curso) + " respostas")

    if curso in quantidade_respostas_dict:
        # Se existir, soma o valor à chave existente
        quantidade_respostas_dict[curso] += quantidade_respostas_curso
    else:
        # Se não existir, cria a chave com o valor desejado
        quantidade_respostas_dict[curso] = quantidade_respostas_curso


print("quantidade total de respostas: " + str(sum(quantidade_respostas_dict.values())))
```

```python
def converter_horario(horario):
    inicio, fim = horario.split(" - ")
    formato = "%H:%M"
    inicio_dt = datetime.strptime(inicio, formato)
    fim_dt = datetime.strptime(fim, formato)
    return inicio_dt, fim_dt
```

```python
def contar(nome_curso,respostas):
    horarios_convertidos = sorted([converter_horario(horario) for horario in respostas], key=lambda x: x[0])
    horarios_ordenados_strings = [f"{inicio.strftime('%H:%M')} - {fim.strftime('%H:%M')}" for inicio, fim in horarios_convertidos]
    dicionario_respostas = {}
    for resposta in horarios_ordenados_strings:
        dicionario_respostas[resposta] = dicionario_respostas.get(resposta,0) +1
    return dicionario_respostas
```

Plotar o Gráfico:

```python
import matplotlib.pyplot as plt

def plotar(nome_curso):
    plot_dict = contar(nome_curso, dados[nome_curso])
    horarios = list(plot_dict.keys())
    contagens = list(plot_dict.values())
    horarios = list(reversed(horarios))
    contagens = list(reversed(contagens))

    fig, ax = plt.subplots()
    ax.set_axisbelow(True)
    # Adiciona as linhas do grid com uma cor mais suave
    ax.grid(axis='x', linestyle='-', alpha=0.3, color='gray')

    # Plota as barras horizontais
    bars = ax.barh(horarios, contagens, color='#2d5868')

    # Adiciona os valores no lado direito das barras
    for valor, categoria in zip(contagens, horarios):
        plt.text(valor + 0.45, horarios.index(categoria), str(valor), ha='left', va='center')

    # Adiciona texto abaixo do gráfico
    plt.figtext(0, -0.05, 'quantidade de votantes: ' + str(quantidade_votantes_dict[nome_curso]) +
                '\nquantidade de respostas: ' + str(sum(contagens)),
                ha='left', va='bottom')

    # Define limites do eixo x
    if max(contagens) < 100:
        plt.xlim(right=max(contagens) + 5)
    else:
        plt.xlim(right=max(contagens) + 10)

    # Adiciona rótulos e título
    plt.xlabel('Respostas')
    plt.ylabel('Horários')
    plt.title('Horários preferidos dos alunos do curso - ' + nome_curso)

    # Exibe o gráfico
    plt.show()
```

```python
for curso in dados:
    plotar(curso)
```

Quantidade Média de horários escolhidos pelos alunos:
(poderia ser visto como a "tolerância" dos alunos quanto aos horários)

```python
def calcular_media(nome_curso):
    quantidade_media_de_respostas = quantidade_respostas_dict[nome_curso]/quantidade_votantes_dict[nome_curso]
    return quantidade_media_de_respostas
```

```python
cursos = list(dados.keys())
print(cursos)
tolerancias = []
for curso in cursos:
    tolerancias.append(calcular_media(curso))


plt.barh(cursos, tolerancias, color='#2d5868')
plt.xlabel('~Tolerância~')
plt.ylabel('Curso')
plt.grid(True)
plt.title('Tolerâncias' )
plt.show()
```

```python
for n in range(0,len(cursos)):
    print(cursos[n] + " : " + str(tolerancias[n]))
```
